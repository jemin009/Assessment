```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/Retail Data.csv")
df
```

# **Retail Sales Analysis**

```python
# 1. View the structure of the dataset (columns, types, missing values).

print(df.info())
print(df.isnull().sum())
```

```python
# 2. What is the shape (rows, columns) of the dataset?

print(df.shape)
```

```python
# 3. Are there any duplicate records?

print(df.duplicated().sum())
```

```python
# 4. Are there any missing or corrupted entries in Ship Date, Order Date, or numeric columns?

print(df[['Order Date','Ship Date']].isnull().sum())

```

```python
# 5. Convert Order Date and Ship Date to datetime.

df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], errors='coerce')
```

```python
# 6.  Check for future or inconsistent shipping dates.

print((df['Ship Date'] < df['Order Date']).sum())
```

```python
# 7. Convert price columns to numeric (remove $ and commas).

cols_to_clean = ['Cost Price', 'Retail Price', 'Profit Margin', 'Sub Total', 'Order Total', 'Shipping Cost']

for col in cols_to_clean:
    if col in df.columns:
        df[col] = df[col].astype(str)  # पहले string में बदलें
        df[col] = df[col].str.replace(r'[\$,]', '', regex=True)  # $, commas हटाएं
        df[col] = pd.to_numeric(df[col], errors='coerce')  # फिर float में बदलें
```

```python
# 8. What are the unique values in Customer Type and Order Priority?

print(df['Customer Type'].unique())
print(df['Order Priority'].unique())
```

```python
# 9. What are the most common shipping modes?

print(df['Ship Mode'].value_counts())
```

```python
# 10. Which cities have the highest number of orders?

print(df['City'].value_counts().head())
```

```python
# 11. What’s the range of order quantities and prices?

print("Order Quantity range:", df['Order Quantity'].min(), "to", df['Order Quantity'].max())

df['Retail Price'] = df['Retail Price'].astype(str).str.replace(r'[\$,]', '', regex=True)
df['Retail Price'] = pd.to_numeric(df['Retail Price'], errors='coerce')

print("Retail Price range:", df['Retail Price'].min(), "to", df['Retail Price'].max())

```

```python
# 12. Create a new column for shipping duration.

df['Ship Duration'] = (df['Ship Date'] - df['Order Date']).dt.days
print(df['Ship Duration'].describe())
```

```python
# 13. Are there any orders with zero or negative total or quantity?

print(df[df['Order Quantity'] <= 0])
```

```python
# 14 Are all discount percentages matching discount dollar amounts?

df['Discount %'] = df['Discount %'].astype(str).str.rstrip('%').astype(float) / 100
df['Discount $'] = df['Discount $'].astype(str).str.replace(r'[\$,]', '', regex=True).astype(float)
df['Retail Price'] = df['Retail Price'].astype(str).str.replace(r'[\$,]', '', regex=True).astype(float)

df['Computed Discount %'] = (df['Discount $'] / df['Retail Price']).round(4)
df['Mismatch'] = df['Computed Discount %'].round(4) != df['Discount %'].round(4)

mismatch_count = df['Mismatch'].sum()
print(f"There are {mismatch_count} orders with mismatched discount percentages.")
```

```python
# 15.  Check for mismatches in total calculation’

df['Profit Margin'] = df['Profit Margin'].astype(str).str.replace(r'[\$,]', '', regex=True)
df['Profit Margin'] = pd.to_numeric(df['Profit Margin'], errors='coerce')

avg_profit_by_container = df.groupby('Product Container')['Profit Margin'].mean()

print("Average Profit Margin by Product Container:")
print(avg_profit_by_container)

top_profitable_container = avg_profit_by_container.idxmax()
print(f"Most profitable product container is: {top_profitable_container}")
```

```python
# 16.  Identify top 5 products by order quantity.

print(df.groupby('Product Name')['Order Quantity'].sum().nlargest(5))
```

```python
# 17. Which Account Manager handled the most revenue?

print(df.groupby('Account Manager')['Order Total'].sum().idxmax())
```

```python
# 18.  What is the average shipping cost by mode?

print(df.groupby('Ship Mode')['Shipping Cost'].mean())
```

```python
# 19. Find the most profitable product.

print(df.groupby('Product Name')['Profit Margin'].sum().idxmax())
```

```python
# 20. Find the most profitable product.

print(df.groupby('Product Name')['Profit Margin'].sum().idxmax())
```

# **Insightful Analysis Questions**

```python
# 1. What is the total revenue generated across all orders?

print(df['Order Total'].sum())
```

```python
# 2. Which customer type generates more revenue?

customer_type = df.groupby('Customer Type')['Order Total'].sum().to_dict()
print(max(customer_type, key=customer_type.get))
```

```python
# 3. How does order priority affect revenue?

order_priority = df.groupby('Order Priority')['Order Total'].sum().to_dict()
print(max(order_priority, key=order_priority.get))
```

```python
# 4. What is the average profit margin by product category?

df['Profit Margin'] = df['Profit Margin'].replace(r'[\$,]', '', regex=True)
df['Profit Margin'] = pd.to_numeric(df['Profit Margin'], errors='coerce')

category_avg = df.groupby('Product Category')['Profit Margin'].mean().to_dict()
print(category_avg)

print(max(category_avg, key=category_avg.get))
```

```python
# 5. What is the most profitable product overall?

profitable_product = df.groupby('Product Name')['Profit Margin'].sum().sort_values(ascending=False).index[0]
print(profitable_product)
```

```python
# 6. How many days does it usually take to ship an order?

df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce', dayfirst=True)
df['Ship Date'] = pd.to_datetime(df['Ship Date'], errors='coerce', dayfirst=True)

df['Shipping Days'] = (df['Ship Date'] - df['Order Date']).dt.days

avg_shipping_days = df['Shipping Days'].mean()
print(avg_shipping_days)
```

```python
# 7. Do longer shipping times impact profit margins?

df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce', dayfirst=True)
df['Ship Date'] = pd.to_datetime(df['Ship Date'], errors='coerce', dayfirst=True)

df['Shipping Days'] = (df['Ship Date'] - df['Order Date']).dt.days

df['Profit Margin'] = df['Profit Margin'].replace(r'[\$,]', '', regex=True)
df['Profit Margin'] = pd.to_numeric(df['Profit Margin'], errors='coerce')

correlation = df[['Shipping Days', 'Profit Margin']].corr().iloc[0, 1]

print("Correlation between shipping days and profit margin:", correlation)
```

```python
# 8. Which city brings in the highest revenue?

top_revenue_city = df.groupby('City')['Order Total'].sum().sort_values(ascending=False).index[0]
print(top_revenue_city)
```

```python
# 9. Which account manager generated the most revenue?

top_account_manager = df.groupby('Account Manager')['Order Total'].sum().sort_values(ascending=False).index[0]
print(top_account_manager)
```

```python
# 10. Which shipping mode is most cost-effective (lowest avg. shipping)?

df['Shipping Cost'] = df['Shipping Cost'].replace(r'[\$,]', '', regex=True)
df['Shipping Cost'] = pd.to_numeric(df['Shipping Cost'], errors='coerce')

avg_shipping_by_mode = df.groupby('Ship Mode')['Shipping Cost'].mean().to_dict()

cheapest_ship_mode = pd.Series(avg_shipping_by_mode).sort_values().index[0]

print(avg_shipping_by_mode)
print(cheapest_ship_mode)
```

```python
# 11. Do higher discounts reduce profits?

if 'Discount %' in df:
    df['Profit Margin'] = df['Profit Margin'].replace(r'[\$,]', '', regex=True)
    df['Profit Margin'] = pd.to_numeric(df['Profit Margin'], errors='coerce')
    corr2 = df['Discount %'].corr(df['Profit Margin'])
else:
    corr2 = np.nan

print("Correlation between Discount Percentage and Profit Margin:", corr2)
```

```python
# 12. Which state has the highest number of orders?

top_order_state = df['State'].value_counts().idxmax()
print(top_order_state)
```

```python
# 13. What is the average discount % across all orders?

df['Discount %'] = df['Discount %'].astype(str)
df['Discount %'] = df['Discount %'].str.rstrip('%').astype(float) / 100.0

avg_discount_pct = df['Discount %'].mean() * 100
print(f"{avg_discount_pct:.2f}%")
```

```python
# 14. What is the average total spend per order?

df['Order Total'] = df['Order Total'].replace(r'[\$,]', '', regex=True)
df['Order Total'] = pd.to_numeric(df['Order Total'], errors='coerce')

avg_order_total = df['Order Total'].mean()
print(f"{avg_order_total:.2f}")
```

```python
# 15. Are certain containers (e.g., Small Box, Wrap Bag) more profitable?

df['Order Total'] = df['Order Total'].replace(r'[\$,]', '', regex=True)
df['Order Total'] = pd.to_numeric(df['Order Total'], errors='coerce')

avg_order_total = df['Order Total'].mean()
print(f"{avg_order_total:.2f}")
```
